```python
#The Collatz Sequence
def collatz(number):
    if number % 2 == 0:
        return number // 2
    else:
        return (3 * number) + 1
        
number = int(input())

i = True
while i:
    if collatz(number) == 1:
        print(collatz(number))
        i = False
    else:
        number = collatz(number)
        print(number)
        collatz(number)
```

```python
#program1
def is_even(num):
    if num % 2 == 0:
        return True
    else:
        return False
print(is_even(4))
print(is_even(0))
print(is_even(245))
print(is_even(343))
print(is_even(39))
```

```python
#program2
def fizzbuss(i):
    for num in range(1, i +1):
        if num % 3 == 0 and num % 5 == 0:
            print('FizzBuzz')
        elif num % 3 == 0:
            print("Fizz")
        elif num % 5 == 0:
            print("Buzz")
        else:
            print(num)
fizzbuss(30)
```
